```python
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets('MNIST_data/', one_hot = True)
```

```python
X = tf.placeholder(dtype = tf.float32, shape = [None, 784])
Y = tf.placeholder(dtype = tf.float32, shape = [None, 10])
drop_prob = tf.placeholder(dtype = tf.float32)
```

```python
W1 = tf.get_variable('W1', initializer = tf.contrib.layers.xavier_initializer(),
                    shape = [784, 588], dtype = tf.float32)     ### get_variable 함수를 이용하여 xavier initializtion 값 사용하기 ###
b1 = tf.Variable(tf.random_normal([588]), dtype = tf.float32)
L1 = tf.nn.relu(tf.matmul(X, W1) + b1)
L1 = tf.nn.dropout(L1, keep_prob = drop_prob)
```

```python
W2 = tf.get_variable('W2', initializer = tf.contrib.layers.xavier_initializer(),
                    shape = [588, 380], dtype = tf.float32)
b2 = tf.Variable(tf.random_normal([380]), dtype = tf.float32)
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)
L2 = tf.nn.dropout(L2, keep_prob = drop_prob)
```

```python
W3 = tf.get_variable('W3', initializer = tf.contrib.layers.xavier_initializer(),
                    shape = [380, 180], dtype = tf.float32)
b3 = tf.Variable(tf.random_normal([180]), dtype = tf.float32)
L3 = tf.nn.relu(tf.matmul(L2, W3) + b3)
L3 = tf.nn.dropout(L3, keep_prob = drop_prob)
```

```python
W4 = tf.get_variable('W4', initializer = tf.contrib.layers.xavier_initializer(),
                    shape = [180, 10], dtype = tf.float32)
b4 = tf.Variable(tf.random_normal([10]), dtype = tf.float32)
L4 = tf.nn.softmax(tf.matmul(L3, W4) + b4)
```

```python
cost = - tf.reduce_mean(tf.reduce_sum(Y * tf.log(L4), axis = 1))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate = 0.001)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
result = tf.equal(tf.arg_max(L4, dimension = 1),
                  tf.arg_max(Y, dimension = 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(result, dtype = tf.float32))
```

```python
epoch = 15
batch_size = 100
for e in range(epoch) :
    iterations = int(mnist.train.num_examples / batch_size)
    for i in range(iterations) :
        batch_x, batch_y = mnist.test.next_batch(batch_size)
        _, cost_val, accuracy_val = \
        sess.run([train, cost, accuracy], feed_dict = {X : batch_x, Y : batch_y, drop_prob : 0.7})
        print(i)
    print(e, cost_val, accuracy_val)
```

```python
sess.run(accuracy, feed_dict = {X : mnist.test.images,
                                Y : mnist.test.labels,
                                drop_prob : 1})
```

```python
################# cnn을 이용한 mnist 분류하기 #########################
```

```python
X = tf.placeholder(dtype = tf.float32, shape = [None, 784])
X_image = tf.reshape(X, [-1, 28, 28, 1])    ### -1 : None과 같음(관측치 수)
Y = tf.placeholder(dtype = tf.float32, shape = [None,10])
```

```python
W1 = tf.Variable(tf.random_normal([3, 3, 1, 16]), dtype = tf.float32)
L1 = tf.nn.conv2d(X_image, W1, strides = [1, 1, 1, 1], padding = 'SAME')    ### padding 옵션에 'SAME' 옵션을 넣게 되면 output과 input의 크기가 같게 된다.
L1 = tf.nn.relu(L1)
L1 = tf.nn.max_pool(L1, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
```

```python
L1
```

```python
W2 = tf.Variable(tf.random_normal([3, 3, 16, 32]), dtype = tf.float32)
L2 = tf.nn.conv2d(L1, W2, strides = [1, 1, 1, 1], padding = 'SAME')
L2 = tf.nn.relu(L2)
L2 = tf.nn.max_pool(L2, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'SAME')
L2_flat = tf.reshape(L2, [-1, 7 * 7 * 32])
```

```python
L2
```

```python
W3 = tf.get_variable('W3', shape = [7 * 7 * 32, 10], initializer = tf.contrib.layers.xavier_initializer(),
                    dtype = tf.float32)
b = tf.Variable(tf.random_normal([10]), dtype = tf.float32)
hypo = tf.nn.softmax(tf.matmul(L2_flat, W3) + b)
```

```python
#cost = -tf.reduce_mean(tf.reduce_sum(Y * tf.log(hypo)))
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = hypo, labels = Y))
optimizer = tf.train.AdamOptimizer(0.001)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
result = tf.equal(tf.arg_max(hypo, 1), tf.arg_max(Y, 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(result, dtype = tf.float32))
```

```python
epoch = 1000
batch_size = 50

for e in range(epoch) :
    iterations = int(mnist.train.num_examples / batch_size)
    for i in range(iterations) :
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        _, cost_val, accuracy_level = sess.run([train, cost, accuracy], feed_dict = {X : batch_x, Y : batch_y})
    print(e, cost_val, accuracy_level)
```
