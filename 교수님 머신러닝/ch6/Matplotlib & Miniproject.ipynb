### Matplotlib
Seaborn과 Plot.ly와 함께 파이썬의 가장 대표적인 데이터 시각화 툴

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(range(100), range(100))   # 하나의 figure 위에 여러 plot을 그릴 수 있음
plt.show()
```

```python
fig = plt.figure()   # figure 반환
fig.set_size_inches(10, 5)

ax_1 = fig.add_subplot(1, 2, 1)   # 하나의 figure 안에 subplot 생성
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c = 'b')
ax_2.plot(X_2, Y_2, c = 'g')
plt.show()
```

#### Set color
rgb color, predefined color 사용 가능

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value + 100 for value in X]

plt.plot(X_1, Y_1, color = '#eeefff')
plt.plot(X_2, Y_2, color = 'r')

plt.show()
```

#### Line style
ls 혹은 linestyle arg로 설정 가능

```python
plt.plot(X_1, Y_1, color = '#eeefff', ls = 'dashed')
plt.plot(X_2, Y_2, color = 'r', linestyle = 'dotted')
plt.title('Two lines')   # title 입력

plt.show()
```

#### Legend

```python
plt.plot(X_1, Y_1, color = '#eeefff', ls = 'dashed', label = 'line 1')
plt.plot(X_2, Y_2, color = 'r', linestyle = 'dotted', label = 'line 2')
plt.legend(shadow = True, fancybox = True, loc = 'lower right')

plt.show()
```

#### Annotation

```python
plt.plot(X_1, Y_1, color = '#eeefff', ls = 'dashed', label = 'line 1')
plt.plot(X_2, Y_2, color = 'r', linestyle = 'dotted', label = 'line 2')
plt.text(50, 70, 'Line_1')
plt.annotate('Line_2', xy = (50, 150), xytext = (20, 175),
            arrowprops = dict(facecolor = 'black', shrink = 0.05))

plt.show()
```

#### Set grid & xylim
Grid와 xy축 한계 지정

```python
plt.plot(X_1, Y_1, color = '#eeefff', ls = 'dashed', label = 'line 1')
plt.plot(X_2, Y_2, color = 'r', linestyle = 'dotted', label = 'line 2')

plt.grid(True, lw = 0.4, ls = '--', c = '.90')
plt.xlim(-100, 200)
plt.ylim(-200, 200)
plt.savefig('test.png', c = 'a')    # figure를 이미지 파일로 저장
plt.show()
```

#### Scatter plot

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0], data_1[:, 1], c = 'b', marker = 'x')
plt.scatter(data_2[:, 0], data_2[:, 1], c = 'r', marker = '^')

plt.show()
```

#### Bubble plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2    # area 변수로 크기를 지정할 수 있음
plt.scatter(x, y, s = area, c = colors, alpha = 0.5)
plt.show()
```

#### Bar chart

```python
data = np.array([[5., 25., 50., 20.],
            [4., 23., 51., 17.],
            [6., 22., 52., 19.]])
X = np.arange(4)

plt.bar(X + 0.00, data[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25)
plt.bar(X + 0.5, data[2], color = 'r', width = 0.25)
plt.xticks(X + 0.25, ('A', 'B', 'C', 'D'))
plt.show()
```

Stacked bar plot

```python
color_list = ['b', 'g', 'r']
data_label = ['A', 'B', 'C']
X = np.arange(data.shape[1])
for i in range(data.shape[0]) :
    plt.bar(X, data[i], bottom = np.sum(data[:i], axis = 0),
           color = color_list[i], label = data_label[i])
plt.legend()
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```

#### Histogram

```python
X = np.random.randn(1000)
plt.hist(X, bins = 10)   # bins : 나누는 구간의 수
plt.show()
```

#### Boxplot

```python
data = np.random.randn(100, 5)
plt.boxplot(data)   # boxplot
plt.show()
```

#### Pandas matplotlib
pandas DataFrame, Series를 이용하여 plotting 가능

```python
df_data = pd.read_csv('boston.csv', header = 'infer')
df_data
```

```python
fig = plt.figure()
ax = []
for i in range(1, 5) :
    ax.append(fig.add_subplot(2, 2, i))  # 2 by 2 figure 만들기
ax[0].scatter(df_data['crim'], df_data['medv'])
ax[1].scatter(df_data['ptratio'], df_data['medv'])
ax[2].scatter(df_data['age'], df_data['medv'])
ax[3].scatter(df_data['nox'], df_data['medv'])
plt.show()
```

```python
fig = plt.figure()
ax = []
for i in range(1, 5) :
    ax.append(fig.add_subplot(2, 2, i))  # 2 by 2 figure 만들기
ax[0].scatter(df_data['crim'], df_data['medv'], label = 'CRIM')
ax[1].scatter(df_data['ptratio'], df_data['medv'], c = 'g')
ax[2].scatter(df_data['age'], df_data['medv'])
ax[3].scatter(df_data['nox'], df_data['medv'])

plt.subplots_adjust(wspace = 0, hspace = 0)   # plot간의 간격 설정
ax[0].legend()
ax[0].set_title('CRIM')

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)
ax = []
for i in range(1, 5) :
    ax.append(fig.add_subplot(2, 2, i))  # 2 by 2 figure 만들기
ax[0].scatter(df_data['crim'], df_data['medv'], label = 'CRIM')
ax[1].scatter(df_data['ptratio'], df_data['medv'], c = 'g')
ax[2].scatter(df_data['age'], df_data['medv'])
ax[3].scatter(df_data['nox'], df_data['medv'])

plt.subplots_adjust(wspace = 0, hspace = 0)   # plot간의 간격 설정
ax[0].legend()
ax[0].set_title('CRIM')

plt.show()
```

```python
df_data.plot()   # pandas 내장 method
```

#### Scaled boxplot

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
scaled_data = std_scaler.fit_transform(df_data.fillna(0))
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.boxplot(scaled_data, labels = df_data.columns)
fig.set_size_inches(18.5, 10.5)
plt.show()
```

#### Scatter matrix

```python
pd.scatter_matrix(df_data, diagonal = 'kde', alpha = 1, figsize = (20, 20))
plt.show()
```

#### correlation plot

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(df_data.corr(), vmin = -1, vmax = 1, interpolation = 'nearest')
fig.colorbar(cax)
fig.set_size_inches(10, 10)
ticks = np.arange(0, 14, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(df_data.columns)
ax.set_yticklabels(df_data.columns)
```
